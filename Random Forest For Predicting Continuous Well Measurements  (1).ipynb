buliding Random Forest Model


```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

Selecting Training and Target Features

```python
X = train_val_df[['RHOB','GR','NPHI','PEF']]
y = train_val_df[['DT']]
```

```python
X_train, X_val, y_train, y_val = train_test_split(X,y, test_size=0.3)
```

```python
X_train
```

Building the model

```python
regr = RandomForestRegressor()
regr.fit(X_train,y_train)
```

```python
y_pred= regr.predict(X_val)
```

```python
y_pred
```

```python
regr.score(X_train,y_train)
```

```python
regr.score(X_val,y_val)
```

```python
df_ypred = pd.DataFrame(y_pred)
```

```python
df_ypred
```

```python
y_val
```

check the prediction results

```python
metrics.mean_absolute_error(y_val, y_pred)
```

```python
mse = metrics.mean_squared_error(y_val, y_pred)
rmse = mse**0.5
rmse
```

in the plot we are comparing the real data against the predicted data

```python
plt.scatter(y_val, y_pred)
plt.xlim(40,140)
plt.ylim(40,140)
plt.plot([40,140],[40,140],'black');

```

Test well prediction

once the model has been tuned , we can apply it to our blind test welll and observe its performance

```python
test_well_x = test_df[['RHOB','GR','NPHI','PEF']]
test_df['TEST_DT'] = regr.predict(test_well_x)
```

```python
plt.scatter(test_df['DT'], test_df['TEST_DT'])
plt.xlim(40,140)
plt.ylim(40,140)
plt.plot([40,140],[40,140],'black');
```
